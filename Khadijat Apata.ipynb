```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
Inventory= pd.read_excel("Sales_transaction .xlsx")
```

```python
Inventory
```

```python
# 1... What is the total revenue generated in the dataset?

Inventory["Revenue"].sum()
```

```python
# 2...  How many total transactions were recorded?

Inventory["TransactionID"].count()
```

```python
# 3.... How many distinct customers made purchases?

Inventory["CustomerID"].nunique()
```

```python
# 4..  What is the total revenue generated from Lagos alone?

Inventory.loc[Inventory["City"]== "Lagos", "Revenue"].sum()
```

```python
# 5... How many units of "Accessories" were sold in Abuja?

Inventory.loc[(Inventory["City"]== "Abuja") & (Inventory["Category"]== "Accessories"),"UnitsSold"].sum()
```

```python
# 6... What is the average discount offered across all transactions?

Inventory["Discount"].mean()
```

```python
# 7.... What was the revenue for February 2024?

Inventory.dtypes
```

```python
Inventory['Date'] = pd.to_datetime(Inventory['Date'])
february_2024_revenue = Inventory[(Inventory['Date'].dt.month == 2) & (Inventory['Date'].dt.year == 2024)]['Revenue'].sum()
```

```python
february_2024_revenue
```

```python
# 8.....
Which month had the highest sales revenue?

Monthly_revenue = Inventory.groupby(Inventory['Date'].dt.to_period('M'))['Revenue'].sum()
highest_revenue_month = Monthly_revenue.idxmax()
highest_revenue_month
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=[10,7])
sns.lineplot(x="Month", y="Revenue",data=Inventory,color="g")
plt.title("BARCHART TO IDENTIFY MONTH WITH THE HIGHEST REVENUE")
```

```python
# 9.... Create a measure to calculate monthly revenue trend

Monthly_trend= Monthly_revenue.reset_index().rename(columns={'Date': 'Month', 'Revenue': 'MonthlyRevenue'})
Monthly_trend
```

```python
# 10.... What is the revenue split between New and Returning customers?
revenue_by_customer_type = Inventory.groupby('CustomerType')['Revenue'].sum()
revenue_by_customer_type
```

```python
plt.figure(figsize=[10,7])
sns.barplot(x="CustomerType", y="Revenue",data=Inventory,color="Purple")
plt.title("BARCHART SHOWING CUSTOMER_TYPE BY REVENUE")
```

```python
# 11....What is the average revenue per customer?

Inventory.groupby('CustomerID')['Revenue'].mean().reset_index()
```

```python
# 12... How many customers bought more than once?
customer_counts = Inventory["CustomerID"].value_counts()
customers_more_than_once = (customer_counts > 1).sum()
customers_more_than_once
```

```python
# 13..... Which product category generated the most revenue?

category_revenue = Inventory.groupby('Category')['Revenue'].sum()
top_category=category_revenue.idxmax()
top_category
```

```python
plt.figure(figsize=[10,7])
sns.barplot(x="Category", y="Revenue",data=Inventory,color="Black")
plt.title("BARCHART TO IDENTIFY CATEGORY WITH THE HIGHEST REVENUE")
```

```python
# 14.... What is the average revenue per unit for Electronics?

electronics_sales = Inventory[Inventory["Category"]== "Electronics"]
avg_revenue_per_unit_electronics = electronics_sales["Revenue"].sum() / electronics_sales["UnitsSold"].sum()
avg_revenue_per_unit_electronics
```

```python
# 15....  Which sales rep generated the highest total revenue?

Top_sales_rep = Inventory.groupby("SalesRep")["Revenue"].sum().idxmax()
Top_sales_rep
```

```python
# 16...  Calculate total units sold by each sales rep.

Inventory.groupby('SalesRep')['UnitsSold'].sum().reset_index()
```

```python
# 17... Calculates revenue from Lagos only if the customer is "New". (New customers only)

Inventory.loc[(Inventory['City']== 'Lagos') & (Inventory['CustomerType']== 'New'),'Revenue'].sum()
```

```python
# 18.... Calculate the total revenue excluding transactions from Port Harcourt

Inventory.loc[Inventory['City']!=("Port Harcourt"), 'Revenue'].sum()
```

```python
# 19....... Calculate average units sold per transaction

Inventory.groupby('TransactionID')['UnitsSold'].mean().reset_index()
```

```python
# 20....Calculate the total discount value across all sales

Inventory['Discount'].sum()
```

```python
Inventory.groupby('SalesRep')['UnitsSold'].mean().reset_index()
```

```python
revenue_split= Inventory.groupby("CustomerType")["Revenue"].sum().reset_index()
```

```python
plt.figure(figsize=(10,8))
revenue_split.plot(kind='pie',x='CustomerType',y='Revenue',color=['blue','green'])
plt.title('Revenue Split by Customer Type')
plt.xlabel('CustomerType')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
